user_id: 사용자의 고유 ip<br>
movie_id : 영화의 고유 ip<br>
rating : 사용자가 영화에 준 평점<br>
timestamp: 평점이 기록된 시간<br>

```python
# 별점 1,2,3 별로 카운트
from pyspark import SparkConf, SparkContext
conf = SparkConf().setMaster('local').setAppName('241204_02_MovieLens')
spark = SparkContext(conf=conf).getOrCreate()
```

```python
import os
directory = os.path.join(os.getcwd(), 'data')
filename = 'u.data'
filepath = os.path.join(directory, filename)
```

```python
# RDD 생성
lines = spark.textFile('file:///'+filepath.replace('\\', '/'))
lines.take(5)
```

```python
header = lines.first()
filtered_lines = lines.filter(lambda row : row != header)
```

```python
filtered_lines.take(5)
```

```python
def parse(row):
    fields = row.split('\t')
    userid = fields[0]
    movie_id  = fields[1]
    rating = fields[2]
    timestamp = fields[3]
    
    return userid, movie_id, rating, timestamp
```

```python
p_lines = filtered_lines.map(parse)
p_lines.collect()
```

```python
rating = p_lines.map(lambda x: (x[2],1))
rating.collect()
```

```python
# 별점별로 필터
rating_filter = rating.filter(lambda x : x[0] in ['1','2','3', '4', '5'])
rating_filter.collect()
```

```python
# 평점이 3이상
rating_filter1 = p_lines.filter(lambda x : x[2] >= '3')
rating_filter1.collect()
```

```python
# 별점별로 합산
rating_sum = rating_filter.reduceByKey(lambda x, y : x + y)
rating_sum.collect()
```

```python
p_lines.count
```

```python
# 평점별 평균
rating_sum_and_count = (
    p_lines.map(lambda x: (x[2], (float(x[2]), 1))).reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))
)
rating_sum_and_count.collect()
```

```python
# 평점별 평균 계산
rating_stats = rating_sum_and_count.mapValues(lambda x: (x[1], x[0] / x[1]))
rating_stats.collect()
```

```python
# 영화id가 1000이상
p_lines.filter(lambda x : int(x[1]) > 1000).collect()
```

```python
# 시간별 평점 합계와 개수 계산
timestamp_sum_and_count = p_lines.map(lambda x: (x[3], (float(x[2]), 1))).reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))
timestamp_sum_and_count.take(5)
```

```python
# 시간별 평균 계산
timestamp_avg_ratings = timestamp_sum_and_count.mapValues(lambda x: (x[1], x[0] / x[1]))
timestamp_avg_ratings.take(5)
```

```python
# 영화별 시간대별 평점 평균 및 개수
# (movie_id, timestamp), (평점 합계, 개수)
movie_time_analysis = p_lines.map(lambda x: ((x[1], x[3]), (float(x[2]), 1))).reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))
movie_time_analysis.take(5)
```

```python
# (movie_id, timestamp), (개수, 평균 평점)
movie_time_analysis_avg = movie_time_analysis.mapValues(lambda x:(x[1], x[0]/x[1]))
movie_time_analysis_avg.take(5)
```
